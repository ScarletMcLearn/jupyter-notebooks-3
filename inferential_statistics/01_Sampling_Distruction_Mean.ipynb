# 01 Sampling Mean, Variability and Central Limit Theorem

The [Central Limit Mean Applet](https://gallery.shinyapps.io/CLT_mean/) let's you play around with creating samples and looking at the sample means.

```python
%matplotlib notebook
import numpy as np
import pandas as pd
from numpy.random import normal
from numpy.random import seed
import matplotlib.pyplot as plt
from sklearn.utils import resample
```

## Define a Population

```python
seed(10)
mu = 50
sigma = 10
pop = pd.DataFrame(normal(mu,sigma,10000),columns=['observation'])
```

```python
pop.plot.hist(bins=30,title='Population Distribution',legend=False,xlim=[0,100],xticks=np.arange(0,110,10))
```

## Create Sampling Distribution

```python
n = 25  # SAMPLE SIZE
n_samples = 10000 # Number of Samples
df = pd.DataFrame(
    [
        resample(pop,n_samples=n,replace=True).mean() for x in range(0,n_samples)
    ])
df.plot.hist(bins=30,legend=False,xlim=[40,60],xticks=np.arange(40,62,2))
```
